```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import math
```

```python
img = cv.imread('C:/Users/ACER/Desktop/02.jpg')
img = np.array(img)/255
# print(img)
plt.figure()
plt.title('original')
plt.imshow(img)

img2 = img+0.5
img2=img2*255
plt.figure()
plt.title('shift')
plt.imshow((img2%255).astype('int'))

img3 = img*1.2
img3=img3*255
plt.figure()
plt.title('multiply')
plt.imshow((img3).astype('int'))
```

```python
mx = np.max(img2)
mn = np.min(img2)
amp = mx-mn
print(f'{mx=}')
print(f'{mn=}')
print(f'{amp=}')
img_restore_shift = ((img2-mn)/amp)*255

mx = np.max(img3)
mn = np.min(img3)
amp = mx-mn
print(f'{mx=}')
print(f'{mn=}')
print(f'{amp=}')
img_restore_multi = ((img3-mn)/amp)*255

plt.figure()
plt.title('restore shift')
plt.imshow(img_restore_shift.astype('int'))

plt.figure()
plt.title('restore multiply')
plt.imshow(img_restore_multi.astype('int'))
```
